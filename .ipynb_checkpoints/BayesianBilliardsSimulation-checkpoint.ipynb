```python
# Alpha Version - Bayesian Billiards 1D

import numpy as np
import matplotlib.pyplot as plt
from scipy.special import beta

# Set the seed for reproducibility
np.random.seed(123)

def thrown_ball(n=3):
    return round(np.random.rand(), n)

def update_counts(white_ball, red_ball, left, overlap, right):
    if red_ball < white_ball:
        left += 1
    elif red_ball > white_ball:
        right += 1
    else:
        overlap += 1

    return left, overlap, right

# Number of iterations
n_iterations = 800

# Initial positions
white_ball = thrown_ball()

# Initialize counters
left = overlap = right = 0

# Lists to store probabilities for plotting
frequentist_probs = []
bayesian_probs = []
white_ball_positions = []

# Simulation loop
for iteration in range(n_iterations):
    # Initial white ball position
    white_ball = white_ball
    
    # Thrown red ball
    red_ball = thrown_ball()

    # Update counts
    left, overlap, right = update_counts(white_ball, red_ball, left, overlap, right)

    # Calculate probabilities
    frequentist = left / (left + right)
    bayesian = (left + 1) / (left + right + 2)

    # Store probabilities for plotting
    frequentist_probs.append(frequentist)
    bayesian_probs.append(bayesian)
    white_ball_positions.append(white_ball)

    # Print results for each iteration
    #print(f"Red ball position: {red_ball}")
    #print(f"Iteration: {iteration} L: {left}, O: {overlap}, R: {right}")
    #print("Frequentist Approach:", frequentist)
    #print("Bayesian Approach:", bayesian)
    #print("---")

# Specify the range of iterations to plot
start_iteration = 100
end_iteration = n_iterations
iterations_to_plot = np.arange(start_iteration, end_iteration + 1)

# Plotting
plt.plot(iterations_to_plot, frequentist_probs[start_iteration-1:end_iteration], color='red', label='Frequentist')
plt.plot(iterations_to_plot, bayesian_probs[start_iteration-1:end_iteration], color='blue', label='Bayesian')
plt.plot(iterations_to_plot, white_ball_positions[start_iteration-1:end_iteration], color='black', label='White Ball')
plt.xlabel('Iteration')
plt.ylabel('Probability')
plt.legend()
plt.title('Frequentist vs Bayesian Probabilities Over Iterations')
plt.show()

# Function to calculate the probability density function (PDF) of the beta distribution
def calculate_PDF(L, R, p):
    numerator = (p ** L) * ((1 - p) ** R)
    denominator = beta(L + 1, R + 1)
    return numerator/denominator

# Parameters
p_values = np.linspace(0, 1, 1000)
L = left
R = right

# Calculate PDF for each p value
pdf_values = [calculate_PDF(L, R, p) for p in p_values]

# Plotting
plt.plot(p_values, pdf_values, label=f'L={L}, R={R}')
plt.xlabel('p')
plt.ylabel('Probability Density')
plt.legend()
plt.title('Probability Density Function Plot')
plt.show()

# Create a grid of p values
p_values = np.linspace(0, 1, 1000)
p_mesh, _ = np.meshgrid(p_values, p_values)

# Calculate PDF for each combination of p values
pdf_values = calculate_PDF(L, R, p_mesh)

# Plotting the heatmap
plt.imshow(pdf_values, extent=(0, 1, 0, 1), origin='lower', aspect='auto', cmap='viridis')
plt.colorbar(label='Probability Density')
# Add a vertical line at the position of the white ball
plt.axvline(x=white_ball, color='red', linestyle='--', label='White Ball Position')
plt.xlabel('p')
plt.ylabel('p')
plt.title('Probability Density Function Heatmap')
plt.show()
```

```python
# Beta Version - Bayesian Billiards 2D

import numpy as np
import matplotlib.pyplot as plt
from scipy.special import beta

# Set the seed for reproducibility
np.random.seed(123)

def thrown_ball(n=3):
    x_position = round(np.random.rand(), n)
    y_position = round(np.random.rand(), n)
    return x_position, y_position

def update_counts(white_position, red_position, left, right, bottom, top, overlap):
    x_white, y_white = white_position
    x_red, y_red = red_position
    
    # Overlap case
    if (x_white == x_red) and (y_white == y_red):
        overlap += 1
    
    # X case
    if x_red < x_white:
        left += 1
    elif x_red > x_white:
        right += 1
    
    # Y case
    if y_red < y_white:
        bottom += 1
    elif y_red > y_white:
        top += 1
    
    return left, right, bottom, top, overlap

# Initialize counts
left_count, right_count, bottom_count, top_count, overlap_count = 0, 0, 0, 0, 0

# Number of balls
num_balls = 100

# Simulate throwing a white ball
white_ball = thrown_ball()

# Initialize 2D arrays for ball positions
red_balls = np.zeros((num_balls, 2))

# Simulate throwing balls and update counts
for i in range(num_balls):
    red_balls[i] = thrown_ball()

    left_count, right_count, bottom_count, top_count, overlap_count = update_counts(
        white_ball, red_balls[i], left_count, right_count, bottom_count, top_count, overlap_count
    )

# Function to calculate the probability density function (PDF) of the beta distribution
def calculate_PDF(A, B, p):
    numerator = (p ** A) * ((1 - p) ** B)
    denominator = beta(A + 1, B + 1)
    return numerator/denominator

# Parameters
p_values = np.linspace(0, 1, 1000)
L = left_count
R = right_count
B = bottom_count
T = top_count

# Calculate PDF for each combination of p values along the x-axis
pdf_x = calculate_PDF(L, R, p_values)

# Calculate PDF for each combination of p values along the y-axis
pdf_y = calculate_PDF(B, T, p_values)

# Create a grid of p values
p_mesh, _ = np.meshgrid(p_values, p_values)

# Create a grid for summing values at each point
grid_prod = np.zeros_like(p_mesh)

# Calculate the sum at each point (x, y)
for i in range(len(p_values)):
    for j in range(len(p_values)):
        grid_prod[i, j] = pdf_x[j] * pdf_y[i]

# White position
x_white, y_white = white_ball

# Plotting the heatmap
plt.imshow(grid_prod, extent=(0, 1, 0, 1), origin='lower', aspect='auto', cmap='viridis')
plt.colorbar(label='Probability Density')
plt.xlabel('p')
plt.ylabel('p')
plt.title('Probability Density Function Heatmap')

# Adding a point at the white_position
plt.scatter(x_white, y_white, color='red', marker='.', label='White Ball')

plt.show()
```
